BASE MODEL
-----------------
➔ Basis-Modell für Regressionsfehler
- schätzen die Zugfestigkeit durch den Median für diese Klasse

```python
import helper
```

```python
zugversuche, kurven, versuche_train, versuche_test = helper.read_data()
```

<b><font color='lightgreen'> Gruppieren nach Klasse, Hinzufügen Mittelwert Zugfestigkeit für diese Klasse

```python
versuche_train['fest_median_group'] = versuche_train.groupby('label')['festigkeit'].transform('median')
versuche_test['fest_median_group']  = versuche_test.groupby('label')['festigkeit'].transform('median')
```

<b><font color='lightgreen'> Vorhersage Trainingsdaten 

```python
festigkeit_train = versuche_train.festigkeit
y_pred_train     = versuche_train.fest_median_group
label_train      = versuche_train.label

helper.plot_pred(y_pred_train, festigkeit_train, label_train, 'Basismodell - Vorhersage Median bei Trainingsdaten', 'mit Linie', 1)
```

<b><font color='lightgreen'> Vorhersage Testdaten 

```python
festigkeit_test = versuche_test.festigkeit
y_pred_test     = versuche_test.fest_median_group
label_test      = versuche_test.label

helper.plot_pred(y_pred_test, festigkeit_test, label_test, 'Basismodell - Vorhersage Median bei Testdaten', 'mit Linie', 1)
```

<b><font color='lightgreen'> Modellbewertung zum Vergleich

```python
helper.get_metric_table_regr(festigkeit_train, y_pred_train, festigkeit_test, y_pred_test, 1)
```
