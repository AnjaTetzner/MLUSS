# OLS 

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
from help_functions import load_stat_feature_pca
```

```python
loader = load_stat_feature_pca(n_components=4)

x_train, x_test, y_train, y_test = loader.load_and_split_regression_data(filter_columns=None)
x_train = sm.add_constant(x_train)
x_test = sm.add_constant(x_test)

model = sm.OLS(y_train, x_train).fit()
y_pred = model.predict(x_test)
print(model.summary())
loader.regression_statistics(y_test, y_pred, rescale = True)
loader.plot_regression_results(y_test, y_pred, rescale = True)
```

```python
residuals = y_test - y_pred
sns.histplot(residuals, kde=True)
plt.title("Verteilung der Residuen")
plt.show()

# Residuen vs. Vorhersagen
plt.scatter(y_pred, residuals)
plt.axhline(0, color="red", linestyle="--")
plt.title("Residuals vs. Predicted")
plt.xlabel("Vorhersagen")
plt.ylabel("Residuen")
plt.show()
```

```python
residuals = y_test - y_pred

stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Q-Q-Plot der Residuen")
plt.show()
```

# Rigde Regression

```python
from sklearn.linear_model import Ridge

loader = load_stat_feature_pca(n_components=3)

x_train, x_test, y_train, y_test = loader.load_and_split_regression_data(filter_columns=None)

ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)

y_pred_ridge = ridge.predict(x_test)

loader.regression_statistics(y_test, y_pred_ridge,rescale = True)
loader.plot_regression_results(y_test, y_pred_ridge,rescale = True)
```

```python
loader = load_stat_feature_pca(n_components=2)

x_train, x_test, y_train, y_test = loader.load_and_split_regression_data(filter_columns=None)

ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)

y_pred_ridge = ridge.predict(x_test)

loader.regression_statistics(y_test, y_pred_ridge,rescale = True)
loader.plot_regression_results(y_test, y_pred_ridge,rescale = True)
```

```python
from sklearn.linear_model import Ridge , RidgeCV
from sklearn.pipeline import Pipeline

loader = load_stat_feature_pca(n_components=4)

x_train, x_test, y_train, y_test = loader.load_and_split_regression_data(filter_columns=None)

ridge_pipeline = Pipeline([
    ("ridge", RidgeCV(alphas=[0.1, 0.5, 1, 5, 10], cv=5))  # RidgeCV mit Cross-Validation
])


ridge_pipeline.fit(x_train, y_train)

y_pred_ridge = ridge_pipeline.predict(x_test)

loader.regression_statistics(y_test, y_pred_ridge,rescale = True)
loader.plot_regression_results(y_test, y_pred_ridge,rescale = True)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso

loader = load_stat_feature_pca(n_components=3)

x_train, x_test, y_train, y_test = loader.load_and_split_regression_data(filter_columns=None)

lasso = Lasso(alpha=1)
lasso.fit(x_train, y_train)

y_pred_lasso = lasso.predict(x_test)

loader.regression_statistics(y_test, y_pred_lasso,rescale = True)
loader.plot_regression_results(y_test, y_pred_lasso,rescale = True)
```

```python
from sklearn.linear_model import Lasso

loader = load_stat_feature_pca(n_components=2)

x_train, x_test, y_train, y_test = loader.load_and_split_regression_data(filter_columns=None)

lasso = Lasso(alpha=1)
lasso.fit(x_train, y_train)

y_pred_lasso = lasso.predict(x_test)

loader.regression_statistics(y_test, y_pred_lasso,rescale = True)
loader.plot_regression_results(y_test, y_pred_lasso,rescale = True)
```

```python
from sklearn.linear_model import LassoCV
from sklearn.pipeline import Pipeline
loader = load_stat_feature_pca(n_components=4)

x_train, x_test, y_train, y_test = loader.load_and_split_regression_data(filter_columns=None)

lasso_pipeline = Pipeline([
    ("lasso", LassoCV(alphas=[0.1, 0.5, 1, 5, 10], cv=5))  
])

lasso_pipeline.fit(x_train, y_train)



y_pred_lasso = lasso_pipeline.predict(x_test)

loader.regression_statistics(y_test, y_pred_lasso,rescale = True)
loader.plot_regression_results(y_test, y_pred_lasso,rescale = True)
```
